# Plots of Loss Functions

## Linear Regression Loss Function Plots

write python code to plot the loss function of a linear regression model

To plot the loss function of a linear regression model, we first need to define a few things:

1. The linear regression model: \( \hat{y} = wx + b \), where \( w \) is the weight and \( b \) is the bias.
2. The squared loss function: \( L = (\hat{y} - y)^2 \), where \( y \) is the ground truth value.
3. A set of ground truth values \( y \) and corresponding input values \( x \).

We will generate a simple dataset with random \( x \) values and corresponding \( y \) values based on a predefined true weight and bias, and then plot the loss for a range of weight and bias values.

Here's a Python script that does this:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple dataset
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
true_w = 3  # true weight
true_b = 4  # true bias
y = true_w * x + true_b + np.random.randn(100, 1)  # y values with some noise

# Loss function
def squared_loss(w, b, x, y):
    y_pred = w * x + b
    return ((y_pred - y) ** 2).mean()

# Range of weights and biases for plotting
w_range = np.linspace(0, 6, 100)
b_range = np.linspace(0, 8, 100)

# Calculate loss values
W, B = np.meshgrid(w_range, b_range)
Loss = np.array([squared_loss(w, b, x, y) for w, b in zip(np.ravel(W), np.ravel(B))])
Loss = Loss.reshape(W.shape)

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, Loss, cmap='viridis')
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Loss')
ax.set_title('Loss Surface for Linear Regression')
plt.show()
```

This code will create a 3D plot showing how the loss varies with different weights and biases for a simple linear regression model. The dataset is synthetic, generated with a preset true weight and bias, and the loss is computed as the mean squared error over this dataset.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple dataset
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
true_w = 3  # true weight
true_b = 4  # true bias
y = true_w * x + true_b + np.random.randn(100, 1)  # y values with some noise

# Loss function
def squared_loss(w, b, x, y):
    y_pred = w * x + b
    return ((y_pred - y) ** 2).mean()

# Range of weights and biases for plotting
w_range = np.linspace(0, 6, 100)
b_range = np.linspace(0, 8, 100)

# Calculate loss values
W, B = np.meshgrid(w_range, b_range)
Loss = np.array([squared_loss(w, b, x, y) for w, b in zip(np.ravel(W), np.ravel(B))])
Loss = Loss.reshape(W.shape)

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, Loss, cmap='viridis')
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Loss')
ax.set_title('Loss Surface for Linear Regression')
plt.show()
````

We will use the same dataset for both plots. To find the best-fitting linear model, we can compute the weight that minimizes the squared loss function, which, for a no-bias model, is the sum of the product of \( x \) and \( y \) divided by the sum of the square of \( x \).

Here's the Python script to do this:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple dataset
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
true_w = 3  # true weight

def squared_loss(w, x, y):

    return ((y_pred - y) ** 2).mean()

# Range of weights for plotting
w_range = np.linspace(0, 6, 100)

# Calculate loss values
loss_values = [squared_loss(w, x, y) for w in w_range]

plt.legend()

plt.tight_layout()
plt.show()
```

This script will produce two subplots:
1. The first subplot shows how the squared loss varies with different weight values.
2. The second subplot shows the original data points and the best-fitting linear model based on the calculated weight.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple dataset
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
true_w = 3  # true weight
y = true_w * x + np.random.randn(100, 1)  # y values with some noise

# Loss function
def squared_loss(w, x, y):
    y_pred = w * x
    return ((y_pred - y) ** 2).mean()

# Range of weights for plotting
w_range = np.linspace(0, 6, 100)

# Calculate loss values
loss_values = [squared_loss(w, x, y) for w in w_range]

# Finding the best weight
best_w = np.sum(x * y) / np.sum(x * x)

# Plotting the loss function
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(w_range, loss_values, color='blue')
plt.xlabel('Weight (w)')
plt.ylabel('Loss')
plt.title('Loss Function for Linear Regression (No Bias)')
plt.grid(True)

# Plotting the data points and the best fitting line
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='red', label='Data points')
plt.plot(x, best_w * x, color='green', label=f'Best fit line (w={best_w:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points and Best Fitting Linear Model')
plt.legend()

plt.tight_layout()
plt.show()
````

## Logistic Regression Loss Function Plots

write python code to plot the loss function of a logistic regression model and allow the user to rotate the plot in 3D

To plot the loss function of a logistic regression model in 3D, we need to consider a model with two parameters, typically the weight \( w \) and the bias \( b \). The logistic regression model predicts the probability of a binary outcome (0 or 1) using the logistic function. The loss function commonly used in logistic regression is the binary cross-entropy loss.

Here's how you can implement this in Python:

1. Generate a simple binary classification dataset.
2. Define the logistic regression model and the binary cross-entropy loss function.
3. Calculate the loss over a range of values for \( w \) and \( b \).
4. Plot the loss in 3D and allow the user to rotate the plot.

Below is the Python code for this:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate a simple binary classification dataset
np.random.seed(0)
x = np.random.randn(100, 1)
y = (x > 0).astype(np.float32)  # 1 if x > 0, else 0

# Logistic regression model
def logistic_regression(w, b, x):
    return 1 / (1 + np.exp(-(w * x + b)))

# Binary cross-entropy loss function
def binary_cross_entropy_loss(w, b, x, y):

b_range = np.linspace(-10, 10, 100)

Loss = Loss.reshape(W.shape)

# Plotting

ax = fig.add_subplot(111, projection='3d')

ax.set_ylabel('Bias (b)')

ax.set_title('Loss Surface for Logistic Regression')

plt.show()
```

This script will create a 3D plot showing the binary cross-entropy loss over a range of weight and bias values for a logistic regression model. Remember, this setup is a simple illustration, and in real-world scenarios, logistic regression models can be more complex and involve multiple features.

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate a simple binary classification dataset
np.random.seed(0)
x = np.random.randn(100, 1)
y = (x > 0).astype(np.float32)  # 1 if x > 0, else 0

# Logistic regression model
def logistic_regression(w, b, x):
    return 1 / (1 + np.exp(-(w * x + b)))

# Binary cross-entropy loss function
def binary_cross_entropy_loss(w, b, x, y):
    predictions = logistic_regression(w, b, x)
    return -np.mean(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))

# Range of weights and biases for plotting
w_range = np.linspace(-10, 10, 100)
b_range = np.linspace(-10, 10, 100)

# Calculate loss values
W, B = np.meshgrid(w_range, b_range)
Loss = np.array([binary_cross_entropy_loss(w, b, x, y) for w, b in zip(np.ravel(W), np.ravel(B))])
Loss = Loss.reshape(W.shape)

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, Loss, cmap='viridis')
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Loss')
ax.set_title('Loss Surface for Logistic Regression')

plt.show()
````

- The loss for a range of weight values is plotted to visualize how the loss changes with the weight of the neuron.
- A simple grid search is used to find the best weight that minimizes the loss.
- The second plot shows the data points and the piecewise linear model (due to the ReLU activation function) that best fits these points, using the weight found in the grid search.

````python
import numpy as np
import matplotlib.pyplot as plt

# ReLU activation function
def relu(x):
    return np.maximum(0, x)

# Single neuron model with ReLU activation and no bias
def neuron_model(w, x):
    return relu(w * x)

# Mean squared error loss function
def mse_loss(w, x, y):
    y_pred = neuron_model(w, x)
    return ((y_pred - y) ** 2).mean()

# Generate a simple dataset
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
true_w = 3  # true weight
y = true_w * relu(x) + np.random.randn(100, 1)  # y values with some noise

# Range of weights for plotting
w_range = np.linspace(-10, 10, 200)

# Calculate loss values
loss_values = [mse_loss(w, x, y) for w in w_range]

# Finding the best weight using a simple grid search
best_w = w_range[np.argmin(loss_values)]

# Plotting
plt.figure(figsize=(12, 5))

# Plotting the loss function
plt.subplot(1, 2, 1)
plt.plot(w_range, loss_values, color='blue')
plt.xlabel('Weight (w)')
plt.ylabel('Loss')
plt.title('Loss Function for a Single Neuron with ReLU Activation (No Bias)')
plt.grid(True)

# Plotting the data points and the best fitting model
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='red', label='Data points')
plt.plot(x, neuron_model(best_w, x), color='green', label=f'Best fit model (w={best_w:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Points and Best Fitting Model')
plt.legend()

plt.tight_layout()
plt.show()
````
